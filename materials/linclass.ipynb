#### Задача 1.
Какое предсказание вернет бинарный линейный классификатор $a(x) = sign(\langle w, x \rangle+w_0)$ для объекта $x=(1, 0, 0, 1, 1)$ при использовании весов $w=(0.1, -0.2, 0.5, -1.1, 0)$ и $w_0=0.35$?

__Решение.__
Запишем скалярное произведение: $\langle w, x \rangle = w_1 x_1 + \dots + w_d x_d$, где $d$ - размерность обоих векторов (число признаков). В нашей задаче $d=5$. Итак, в скобках получится значение $0.1\cdot1 - 0.2\cdot0 + 0.5\cdot 0 -1.1 \cdot 1 + 0 \cdot 1 + 0.3 = -0.7$. Его знак отрицательный, значит $a(x) = -1$.

#### Задача 2.
Визуализируйте разделящую поверхность классификатора $a(x) = sign(\langle w, x \rangle+w_0)$ для $w=(-1, 2)$, $w_0=0.5$, задача бинарной классификации с двумя признаками.

__Решение.__

Предсказания в линейной классификации выполняются по формуле $a(x) = sign(\langle w, x\rangle +w_0)$, то есть класс +1, если выражение в скобках больше 0, и -1, если выражение в скобках меньше 0. Если выражение в скобках равно 0, считаем, что отказываемся от классификации или выбираем случайный класс (на практике такая ситуация встречается очень редко). Соответственно, мы можем сделать такое предсказание в каждой точке признакового пространства, то есть для любого возможного объекта, и получить области классов +1 и -1. Разделяющей границей этих двух областей будет прямая, поэтому классификация линейная.

Разберемся, почему разделяющей границей будет прямая. Вспомним, что уравнение $w_1 x_1 + w_2 x_2 + w_0 = 0$ задает прямую на плоскости в координатах $x_1-x_2$ (мы обычно обозначаем эти координаты $d_1-d_2$, чтобы не путать $x_1$ - первый признак или первый объект в выборке; в данном контексте - первый признак). Все точки $x=(x_1, x_2)$, для которых $w_1 x_1 + w_2 x_2 + w_0 > 0$, находятся с одной стороны от прямой, а все точки, для которых $w_1 x_1 + w_2 x_2 + w_0 < 0$, с другой стороны от прямой. А эти два условия как раз и проверяются в бинарном линейной классификаторе.

В нашем случае прямая задается уравнением $-x_1 + 2 x_2 + 0.5 = 0$. Построим ее по точкам: при $x_1=0$ выполнено $2 x_2+0.5=0$, то есть $x_2=-1/4$; при $x_1=1$ выполнено $2 x_2 -0.5=0$, то есть $x_2=1/4$. Проводим прямую через точки $(0, -1/4)$ и $(1, 1/4)$. Теперь выбираем любую точку, не принадлежащую прямой, например $(0, 1/4)$, она обозначена P на изображении. Проверяем, какой будет знак выражения $-x_1 + 2 x_2 + 0.5$: $0 + 0.5 + 0.5 = 1 > 0$. Значит, в полуплоскости, где находится эта точка, мы предсказываем класс +1 (обозначен желтым), а в другой - класс -1 (обозначен зеленым).

#### Задача 3.
Изобразите решающее правило логистической регрессии по одному признаку с его весом 2 и свободным членом 0.6 в осях первый признак - вероятность класса +1. Укажите, для каких $x$
 будет предсказан класс +1, для каких - класс -1, если бинаризация выполняется по порогу 0.5 на вероятность класса +1.
 
__Решение.__

В логистической регрессии предсказывается вероятность класса +1 для объекта $x$ по формуле:

$$P(y=+1|x) = \sigma(\langle w, x\rangle +w_0) \in (0, 1),$$

где $\sigma(z) = \frac 1 {1+\exp(-z)}$ - функция сигмоиды, монотонно преобразующая числовую ось в отрезок (0, 1).

В нашем случае признак только один, и $\langle w, x\rangle + w_0 = w_1 x_1 + w_0 = 2 x_1 + 0.6$.

Итак, нам нужно построить график функции $P(y=+1|x) = \sigma(2 x_1 + 0.6)$. Это функция от одной переменной $x_1$. Чтобы ее построить, можно взять график сигмоиды (оранжевая линия), сжать его в два раза (т. к. $2 x_1$; зеленая линия) и сдвинуть на 0.3 влево (т. к. $2 x_1 + 0.6 = 2(x_1+0.3)$). При этом область значений функции не изменится и будет равна $(0, 1)$.

Для всех абсцисс $x_1$, для которых значение изображенной функции больше 0.5 (вероятность класса +1 больше 0.5), предсказывается класс +1, аналогично для класса -1. В данной задаче получается, что для $x_1 < -0.3$ предсказывается класс -1 (на изображении ниже отмечен зеленым цветом), для $x_1 > -0.3$ предсказывается класс +1 (фиолетовый цвет). Порог -0.3 можно увидеть по графику, а можно вычислить. Чтобы вычислить порог -0.3, решаем уравнение $2 x_1 + 0.6=0$, потому что в нуле сигмоида равна 0.5 - порогу на вероятность класса +1.

#### Задача 4\* (повышенной сложности).
В бинарных классификаторах можно использовать разные пороги бинаризации $T$: классификатор $a(x) = sign(\langle w, x\rangle +w_0 - T)$ возвращает класс +1, если выражение $\langle w, x\rangle +w_0$ больше $T$, и -1, если выражение меньше $T$. Аналогично можно менять пороги $P$ в логистической регрессии: если вероятность класса +1 больше порога $P$, предсказывается класс +1, если меньше - класс -1. Как связаны пороги $T$ и $P$?

__Решение.__
В логистической регрессии предсказывается вероятность класса +1 для объекта $x$ по формуле:

$$P(y=+1|x) = \sigma(\langle w, x\rangle +w_0) \in (0, 1)$$

Эти вероятности можно по-разному использовать на практике, например, по вероятности можно оценивать уверенность классификатора в предсказании и, если он не уверен, отказываться от классификации. Такое поведение классификатора может быть полезно в некоторых приложениях с большой ценой ошибки, например, в медицине.

Если же мы хотим предсказать бинарный класс +1 или -1, мы сравниваем вероятность класса $+1$ с порогом: 

$$a(x) = sign (\sigma(\langle w, x\rangle +w_0)-P).$$

Функция $\sigma$ монотонна, поэтому выражение в скобках можно преобразовать, чтобы получить порог на $\langle w, x\rangle +w_0$. Мы сделаем это, выполним цепочку преобразований неравенств и вспомнив определение сигмоиды $\sigma(z) = \frac 1 {1+\exp(-z)}$:

$$\sigma(\langle w, x\rangle +w_0) > P$$

$$\frac 1 {1 + \exp (-\langle w, x\rangle -w_0)} > P$$

$$1 + \exp (-\langle w, x\rangle -w_0) < \frac 1 P$$

$$\exp (-\langle w, x\rangle -w_0) < \frac 1 P - 1$$

$$-\langle w, x\rangle -w_0 < \log \biggl(\frac 1 P  - 1 \biggr)$$

$$\langle w, x\rangle + w_0 > -\log \biggl(\frac 1 P - 1\biggr)$$

Правая часть последнего неравенства и есть искомый порог $T = -\log \bigl( \frac 1 P - 1 \bigr)$.

Проверим себя, подставив $P=0.5$: $T= - \log 1 = 0$.

__Обратите внимание:__ аналогично линейной регрессии, в линейной классификации можно добавить единичный константный признак в матрицу объекты-признаки $X$. В этом случае свободный член $w_0$ в формулах не пишется, так как входит в общий вектор весов $w$.

#### Задача 5.

Предположим, что на определенной обучающей выборке два класса можно линейно разделить в признаковом пространстве. В этом случае оптимальные веса логистической регрессии будут бесконечными. Почему?

__Решение.__

Вспомним, как обучаются бинарные линейные классификаторы. Мы вводим понятие отступа на одном объекте: $M_i = y_i \langle w, x\rangle +w_0$. Если $M_i > 0$, объект классифицируется правильно (знаки $\langle w, x\rangle +w_0$ и класса совпадают), если $M_i < 0$ - неправильно. По модулю $|M_i|$ можно понять, насколько классификатор уверен в своем предсказании: если $|M_i|\approx 0$, небольшое изменение весов может привести к изменению предсказания, то есть классификатор не уверен. Если же $|M_i| >> 0$, классификатор не станет менять свое предсказание при небольшом изменении весов.

Чтобы найти оптимальные для выборки веса, мы вводим функцию потерь $L(M)$ и минимизируем сумму значений функции потерь на объектах обучающей выборки по весам:

$$\sum_{i=1}^\ell L(M_i) \rightarrow \min_{w, w_0}$$

В логистической регрессии используется логистическая функция потерь: $L(M_i) =  \bigl(1+\exp(-M_i) \bigr)$. Она изображена на рисунке ниже. 

<div>
<img src="logistic.png" width="350"/>
</div>

Эта функция потерь сильно штрафует $M_i < 0$, немного штрафует $|M_i|\approx 0$ и почти не штрафует $|M_i| >> 0$. 

Теперь вспомним условие: два класса можно линейно разделить в признаковом пространстве. Это означает, что существуют веса $\tilde w$, для которых все объекты обучающей выборки будут классифицироваться правильно. Это означает, что для всех объектов $M_i > 0$. Но функция $L(M_i)$ монотонно убывает в бесконечность, а значит, чем больше $M_i$ мы возьмем, тем меньше будет значения функционала качества $Q(w)$. Например, возьмем веса $100 \tilde w$, $100 w_0$. При таких весах все отступы $M_i$ увеличатся в 100 раз, а значение $Q(w)$ уменьшится. В итоге оптимальные веса будут бесконечными.

#### Задача 6\* (повышенной сложности).
Выведите логистическую функцию потерь на основе метода максимального правдоподобия для обучения логистической регрессии.

__Решение.__

У логистической регрессии есть две постановки: (А) через предсказание метки -1/+1 и (Б) через предсказание вероятности класса +1. Обе постановки приводят к одному и тому же функционалу качества и к одному и тому же методу обучения весов.

Для упрощения обозначений введем $b(x) = \langle w, x\rangle +w_0$.

Предсказания в постановке (А) выполняются по формуле $a(x) = sign(b(x)) \in \{-1, +1\}$, в постановке (Б) - по формуле $P(y=+1|x) = \sigma(b(x)) \in (0, 1)$. Один набор весов можно использовать в формуле (А), чтобы предсказать метку класса, и в формуле (Б), чтобы предсказать вероятность.

В постановке (Б) обучение логично проводить методом максимального правдоподобия: максимизируем вероятность наблюдаемых в данных меток по весам модели.

$$Q(w) = \sum_{i=1}^\ell \bigl ( [y_i=+1] \log P(y_i=+1|x_i) + [y_i=-1] \log (1-P(y_i=-1|x_i)) \bigr) \rightarrow \max_{w, w_0}$$

Здесь зависимость от вектора весов $w$ "прячется" в вероятности $P(y_i=+1|x_i)$. Подставим формулу с сигмоидой:

$$Q(w) = \sum_{i=1}^\ell \bigl ( [y_i=+1] \log \sigma(b(x_i)) + [y_i=-1] \log (1-\sigma(b(x_i))) \bigr).$$

Преобразуем $\log \sigma(b(x_i))$:

$$\log \sigma(b(x_i)) = \log \frac 1 {1+\exp(-b(x_i))} = 
- \log \bigl(1+\exp(-b(x_i))\bigr).$$

Преобразуем $\log (1-\sigma(b(x_i)))$. Для этого сначала преобразуем выражение в скобках, приведя к общему знаменателю:

$$1-\sigma(b(x_i)) = 1 - \frac 1 {1+\exp(-b(x_i))} = 
\frac{1+\exp(-b(x_i))-1} {1+\exp(-b(x_i))}  = \frac{\exp(-b(x_i))} {1+\exp(-b(x_i))} = \frac 1 {1+\exp(b(x_i))} = \sigma(-b(x_i)).$$

В предпоследнем преобразовании мы воспользовались основным свойством дроби и разделили числитель и знаменатель на $b(x_i)$. Теперь по аналогии с первым логарифмом:

$$ \log \sigma(-b(x_i)) = \log \frac 1 {1+\exp(b(x_i))} = 
- \log \bigl(1+\exp(b(x_i))\bigr). $$

Получилось выражение:

$$Q(w) = \sum_{i=1}^\ell \bigl ( - [y_i=+1] \log \bigl(1+\exp(-b(x_i))\bigr) - [y_i=-1] \log \bigl(1+\exp(b(x_i))\bigr) \bigr).$$

Слагаемые в обоих выражениях очень похожи, и их можно объединить в одном выражении $- \log \bigl(1+\exp(-y_i b(x_i)) \bigr) $. Для проверки можно подставить $y=-1$ и $y=+1$. В итоге получим функционал

$$Q(w) = - \sum_{i=1}^\ell \log \bigl(1+\exp(-y_i b(x_i)) \bigr)  \rightarrow \max_{w, w_0}$$

Здесь зависимость от $w$ заключена в $b(x_i)$.

Теперь, если вспомнить определение отступа на одном объекте $M_i = y_i \langle w, x\rangle +w_0 = y_i b(x_i)$, можно увидеть, что мы получили логистическую функцию потерь:

$$L(M_i) =  \bigl(1+\exp(-M_i) \bigr),$$

а функционал $Q(w)$ - сумма функций потерь по отдельным объектам обучающей выборки. Именно по такому алгоритму - оптимизация суммы функций потерь, зависящих от отступа - происходит обучение бинарных линейных классификаторов.

В итоге, мы начали от обучения в постановке (Б) и пришли к обучению в постановке (А). При этом в обеих постановках мы оптимизируем одну и ту же функцию, только по-разному записанную. Значит, в итоге мы получим одни и те же веса, в какой бы постановке мы ни обучали логистическую регрессию.

#### Важности и веса признаков
Большинство алгоритмов умеют так или иначе оценивать важности признаков. 
В линейной модели в качестве важностей можно рассматривать веса признаков. Они хранятся в атрибуте coef\_ и появляются, конечно, только после вызова процедуры обучения.

```python
clf_lr.coef_
```

__Задание.__ Оформить веса признаков в виде датафрейма: первый столбец - имя признака, второй столбец - вес, и отсортировать датафрейм по увеличению веса.

__Решение:__

```python
pd.DataFrame({"feature":tab.columns[:-2],
             "weight":clf_lr.coef_[0]}).sort_values("weight")
```

Признаки отсортировались по логичным критериям: плата за съем жилья, число кредитов, заемщик-иностранец уменьшают шанс получить кредит; наличие собственности, машины, работы, счета в банке - увеличивают шансы.

Впрочем, некоторые признаки отсортировались менее логично: например, наличие поручителя тоже голосует в "минус", хотя и с маленьким весом.

__Обратите внимание:__ интерпретировать величину весов можно, только если данные отнормированы. Иначе модуль веса будет зависеть от масштаба признака.

#### Работа с категориальным признаком

Применим метод one-hot-encoding к переменной "цель получения кредита", чобы включить ее в модель. Для этого воспользуемся функцией pd.get_dummies

```python
tab_ohe = pd.get_dummies(tab, "purpose")
```

```python
tab_ohe.head()
```

```python
# удаляем целевую переменную с помощью метода drop
X_ohe = tab_ohe.drop("target", axis=1)
# axis=1 показывает, что мы отим удалить столбец, а не строку (axis=0)
```

```python
X_train_ohe, X_test_ohe, y_train, y_test  = train_test_split(X_ohe, y, \
                                                     test_size=0.3,\
                                                     shuffle=True,
                                                     random_state=0)
```

```python
scaler = StandardScaler()
X_train_ohe = scaler.fit_transform(X_train_ohe)
X_test_ohe = scaler.transform(X_test_ohe)
```

Благодаря фиксации random_state=0 мы получаем одно и то же разделение.

Оформим обучение классификатора и подсчет качества в виде функции:

```python
def get_accuracy(clf):
    clf.fit(X_train_ohe, y_train)
    y_pred = clf.predict(X_test_ohe)
    return accuracy_score(y_test, y_pred)
```

Качество с новым признаком повысилось.

А что, если мы просто пронумеруем категории? Для этого воспользуемся классом LabelEncoder:

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
tab["purpose"] = le.fit_transform(tab["purpose"])
```

```python
X_le = tab[tab.columns[:-1]]
```

```python
X_train_le, X_test_le, y_train, y_test  = train_test_split(X_le, y, \
                                                     test_size=0.3,\
                                                     shuffle=True,
                                                     random_state=0)
```

```python
scaler = StandardScaler()
X_train_le = scaler.fit_transform(X_train_le)
X_test_le = scaler.transform(X_test_le)
```

```python
def get_accuracy(clf):
    clf.fit(X_train_le, y_train)
    y_pred = clf.predict(X_test_le)
    return accuracy_score(y_test, y_pred)
```

```python
print(get_accuracy(LogisticRegression()))
```

При использовании LabelEncoding качество работы классификатора не улучшилось по сравнению с отсутствием категориального признака, что логично.
